# data visualisation 101 in python

```python
# import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#read in data 
demographic_data = pd.read_csv("data/life_expectancy_and_income.csv")
```

```python
demo_data_2019 = (
    demographic_data
    .loc[demographic_data.year == 2019]
    .copy()
)
```

```python
#life expectancy vs income in 2019
plt.scatter(
    x="income_per_person",
    y="life_expectancy",
    data=demo_data_2019
)

#axis labels
plt.xlabel("Income per person($)")
plt.ylabel("Life expectancy (years)")

#title 
plt.title("Life expectancy vs income in 2019")

#set limits
plt.xlim(0, 80000)
plt.ylim(0, 100)

#controling tick marks
plt.yticks(list(range(0, 110, 10)))

#turns off rendering code around plot
plt.show()
```

```python
#histogram of income per person in 2019
plt.hist(x="income_per_person",
        data=demo_data_2019)

plt.xlabel("income per person ($)")
plt.ylabel("count")
plt.title("distribution of income per person")

plt.show
```

```python
#boxplot of income per person in 2019
plt.boxplot(x="income_per_person", data=demo_data_2019)
plt.ylabel("income per person ($)")
plt.title("Distribution of income per person 2019")
plt.show()
```

## plotting with seaborn

```python
#life expectancy vs income in 2019 
sns.scatterplot(x="income_per_person",
                y="life_expectancy",
                data=demo_data_2019
               )

plt.xlabel("Income per person ($)")
plt.ylabel("Life expectancy (years)")

plt.show()
```

```python
sns.histplot(x="income_per_person", 
            data=demo_data_2019)
```

```python
sns.boxplot(x="income_per_person",
           data=demo_data_2019)
```

```python
#fertility over time in japan 
japan_data = (demographic_data.loc[demographic_data.country == "Japan"].copy())
```

```python
sns.lineplot(x="year", y="fertility_rate", data=japan_data)
```

```python
#income per person for the g7 nations in 2019
g7 = ["Canada", "Italy", "France", "Germany", "United Kingdom", "United States", "Japan"]
```

```python
g7_2019 = (
    demo_data_2019.
    loc[demo_data_2019.country.isin(g7)]
    .sort_values("income_per_person")
    .copy()
)
```

```python
sns.barplot(x="country",
            y="income_per_person",
            data=g7_2019, 
            color="steelblue")

plt.xticks(rotation=45)

plt.show()
```

```python
#axis flip 
sns.barplot(y="country",
            x="income_per_person",
            data=g7_2019, orient="h")

plt.show()
```

```python
#task
occupation_prestige = pd.read_csv("data/prestige_occupation.csv")
sns.scatterplot(x="prestige",
               y="education",
               data=occupation_prestige)

plt.show()
```

```python
min_score = (
    occupation_prestige.loc[
        occupation_prestige
        .groupby("type")
        .prestige
        .idxmin()
        .sort_values()]
    .copy()
)
```

```python
sns.barplot(x="type",
           y="prestige",
           data=min_score)
```

## colours and themes

```python
sns.scatterplot(x="income_per_person", 
               y="life_expectancy",
               data=demo_data_2019,
               color="firebrick")
```

```python
sns.scatterplot(x="income_per_person", 
               y="life_expectancy",
               data=demo_data_2019,
                #colour depends on data
               hue="fertility_rate",
               #set palette
               palette="viridis_r",
               #size: depends on data
               size = "fertility_rate")
```

```python
#categorical colours
#draw attention to countries with high fertility rate 

#make high fertility rate column

import numpy as np
```

```python
demo_data_2019["high_fertility_rate"] = np.where(
    demo_data_2019.fertility_rate > 5, True, False)

demo_data_2019.head()
```

```python
sns.scatterplot(
    x="income_per_person", 
    y="life_expectancy",
    data=demo_data_2019,
    #colour depends on data
    hue="high_fertility_rate",
    #categorical palette
    palette="colorblind"
)
```

```python
sns.scatterplot(
    x="income_per_person", 
    y="life_expectancy",
    data=demo_data_2019,
    #colour depends on data
    hue="high_fertility_rate",
    #our own palette
    palette={True: "tomato", False: "seagreen"},
    #change the shape of our points
    style="high_fertility_rate"
)
```

```python
#task
sns.scatterplot(
    x="prestige",
    y="education",
    data=occupation_prestige,
    hue="type"
)
```

```python
sns.scatterplot(
    x="prestige",
    y="education",
    data=occupation_prestige,
    hue="income"
)
```
